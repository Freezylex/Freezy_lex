```python
import pandas as pd
import requests
```

```python
url = 'https://www.cbr.ru/hd_base/zcyc_params/?UniDbQuery.Posted=True&UniDbQuery.From=02.03.2015&UniDbQuery.To=19.02.2016'
answer = requests.get(url)
```

```python
data = pd.read_html(answer.content)[0]
data = data[::-1]
data = data.droplevel(0, axis=1)
data['Дата'] = data['Дата'].apply(lambda x: '-'.join(x.split('.')[::-1]))
data['Дата'] = pd.to_datetime(data['Дата'])
data.set_index('Дата', inplace=True)
data.columns = [float(x.replace(',', '.')) for x in data.columns]
data = data.astype('int64')
data = data/10 ** 4
data.index.rename('Date', inplace=True)
data
```

```python
data[0] = data[0.25]
data = data[[data.columns[-1]] +  list(data.columns[:-1])]
data
```

```python
data.to_excel('cbr_rates.xlsx')
```
