```python
import pandas as pd
from tqdm import tqdm
import numpy as np
import os
import gc
import json
```

```python
a_file = open("bonds_info.json", "r")
data = json.load(a_file)
data
```

```python
list_of_keys = list(data.keys())
list_of_keys
```

```python
start_path = '/Volumes/LaCie/Trade Data/'
files = os.listdir(start_path)
files = [file for file in files if not os.path.isfile(file)]
for file in tqdm(files):
    data = pd.read_csv(start_path + file + '/' + file + '.txt')
    vals = data['SECCODE'].unique()
    vals = [val for val in vals if len(val) > 7 and val[:-1] in list_of_keys]
    bonds = data[data['SECCODE'].isin(vals)]
    del data
    bonds.to_csv(path_or_buf='/Users/freezylex/Desktop/курсач/days/' + file + '.csv')
    gc.collect()
```
