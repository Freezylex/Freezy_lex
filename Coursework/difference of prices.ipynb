# Confidence intervals

```python
hypothesis = pd.DataFrame(columns=['t_stat', 'lower_95', 'higher_95', 'p_value'],
                         index = difference.columns.values)
for column in difference.columns:
    a = difference[column].values
    lower, higher = st.t.interval(0.95, len(a)-1, loc=np.mean(a), scale=st.sem(a))
    t_stat = np.mean(a)/np.std(a, ddof=1)
    p_value = 2*(1 - st.t.cdf(np.abs(t_stat), df = len(a)-1))
    hypothesis.loc[column, 't_stat'] = t_stat
    hypothesis.loc[column, 'lower_95'] = lower
    hypothesis.loc[column, 'higher_95'] = higher
    hypothesis.loc[column, 'p_value'] = p_value
```
