```python
all_data = pd.DataFrame(columns=tickers)
first = True
for file in files:
    data = pd.read_excel(path+'/'+file)
    if first:
        data['Дата'] = pd.to_datetime(data['Дата'], format='%Y-%m-%d')
        all_data['Date'] = data['Дата']
        first = False
    else:
        pass
    bond = 'SU' + str(file[:5]) + 'RMFS'
    red_date = pd.to_datetime(data_bonds[bond]['redemption_date'], format='%d.%m.%Y')
    till_red = (red_date - all_data['Date'])/timedelta(days=365)
    all_data[bond] = till_red
```
